#### Clase Asistencia con 3  atributos (nombre, fecha, estado)
#### 2 funciones leerdatos(), mostrardatos()
#### Que sea interactivo --> Los input
#### 2 objetos (asistencia1, asistencia2).

```python
# Clase Asistencia con atributos: nombre, fecha, estado
# Funciones: leerdatos(), mostrardatos()
class Asistencia:
    def __init__(self, nombre, fecha, estado):
        self.nombre = nombre
        self.fecha = fecha
        self.estado = estado

    def leerdatos(self):
        self.nombre = input("Ingrese el nombre del alumno: ")
        self.fecha = input("Ingrese la fecha (DD/MM/AAAA): ")
        self.estado = input("Ingrese el estado (Asistió/Faltó): ")
        return self.nombre, self.fecha, self.estado

    def mostrardatos(self):
        print("Nombre del alumno:", self.nombre)
        print("Fecha de clase:   ", self.fecha)
        print("Estado:           ", self.estado)
        return self.nombre, self.fecha, self.estado
    # Fin de la clase Asistencia

# Creando los objetos asistencia1 y asistencia2
print("Registro de Asistencia número 1")
asistencia1 = Asistencia("", "", "") # Creando objeto asistencia1
asistencia1.leerdatos()
asistencia1.mostrardatos()

print("-------------------")

print("Registro de Asistencia número 2")
asistencia2 = Asistencia("", "", "") # Creando objeto asistencia2
asistencia2.leerdatos()
asistencia2.mostrardatos()

print("-------------------")
print("Software desarrollado por Samuel Martinez NL = 35+3=38")
```
